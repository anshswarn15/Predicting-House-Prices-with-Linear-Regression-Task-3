# Data Preparation

```python
varlist =  ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
def binary_map(x):
    return x.map({'yes': 1, "no": 0})
housing[varlist] = housing[varlist].apply(binary_map)
```

```python
housing.head()
```

```python
status = pd.get_dummies(housing['furnishingstatus'])
```

```python
status.head()
```

***From above, drop the furnished column, as the type of furnishing can be identified with just the last two columns where 
00 will correspond to furnished;
01 will correspond to unfurnished;
10 will correspond to semi-furnished***

```python
status = pd.get_dummies(housing['furnishingstatus'], drop_first = True)
```

```python
housing = pd.concat([housing, status], axis = 1)
```

```python
housing.head()
```

```python
housing.drop(['furnishingstatus'], axis = 1, inplace = True)
```

```python
housing.head()
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
df_train, df_test = train_test_split(housing, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20, 16))
sns.heatmap(df_train.corr(), annot=True, cmap="YlGnBu", annot_kws={"size": 12}, fmt=".2f")
plt.show()
```

***From above heatmap, We can say that area seems to the correlated to price the most.***

```python
y_train = df_train.pop('price')
X_train = df_train
```

# Model Building(LinearRegression function)

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
lm = LinearRegression()

# Initialize RFE with the linear model and the number of features to select
rfe = RFE(estimator=lm, n_features_to_select=6)

# Fit the RFE model
rfe = rfe.fit(X_train, y_train)

# Display the selected features
print("Selected features:", rfe.support_)
print("Feature ranking:", rfe.ranking_)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

***Building model using statsmodel, for the detailed statistics***

```python
X_train_rfe = X_train[col]
```

```python
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit() 
```

```python
print(lm.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

***Residual Analysis of the train data***

```python
y_train_price = lm.predict(X_train_rfe)
```

```python
res = (y_train_price - y_train)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  
plt.xlabel('Errors', fontsize = 18)                         
```

```python
plt.scatter(y_train,res)
plt.show()
```

***Model Evaluation***

```python
num_vars = ['area','stories', 'bathrooms', 'airconditioning', 'prefarea','parking','price']
```

```python
df_test[num_vars] = scaler.fit_transform(df_test[num_vars])
```

```python
y_test = df_test.pop('price')
X_test = df_test
```

```python
X_test = sm.add_constant(X_test)
```

```python
X_test_rfe = X_test[X_train_rfe.columns]
```

```python
y_pred = lm.predict(X_test_rfe)
```

```python
from sklearn.metrics import r2_score 
r2_score(y_test, y_pred)
```

```python
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)               
plt.xlabel('y_test', fontsize=18)                          
plt.ylabel('y_pred', fontsize=16)
```

***From Model, We can say that the equation of our best fitted line is:***

***price = 0.35area + 0.20bathrooms + 0.19stories + 0.10airconditioning + 0.10parking + 0.11prefarea***

# Conclusion for Predicting House Prices with Linear Regression

	The models best-fit equation was:
    price = 0.35area + 0.20bathrooms + 0.19stories + 0.10airconditioning + 0.10parking + 0.11prefarea
	Area was the most influential feature on house prices.The linear regression model for predicting house prices indicates that key features such as area, number of bedrooms, bathrooms, stories, main road access, guestroom availability, basement, air conditioning, parking spaces, preference area, and furnishing status significantly impact house prices.
